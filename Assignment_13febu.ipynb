### Ques-No.4
Why LookupError class is used? Explain with an example KeyError and IndexError.

```python
>Lookup Error acts as a base class for the exceptions that occur
when a key or index used on a mapping or sequence of a list/dictionary is invalid or does not exists.
```

```python
try:
    l=[1,2,3,4,5]
    print(l[6])
except IndexError as e:
    print(e)
```

### Ques-No.6
List down some best practices for exception handling in python.

```python
>> Avoid raising generic exceptions because if you do so,
then all other more specific exceptions have to be caught also.
Hence,
the best practice is to raise the most specific exception close to your problem.
```

```python
try:
    10/0
except FileNotFoundError as e:
    logging.error("I am trying to handling the file not found error".format(e))
except AttributeError as e:
    logging.error("I am trying to handling the attribute error".format(e))
    
except ZeroDivisionError as e:
    logging.error("I am trying to handling the zero division by error".format(e))
        
```
